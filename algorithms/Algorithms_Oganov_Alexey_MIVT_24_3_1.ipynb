1. Сделайте функцию, которая параметром будет принимать список и
удалять из него все дубли.

```python
def no_duplicates(input_list):
    seen = set()
    unique = []
    for item in input_list:
        if item not in seen:
            unique.append(item)
            seen.add(item)
    return unique

original = [22, 23, 22, 64, 64, 65, 71]
unique = no_duplicates(original)
print(unique)
```

6. Сделайте функцию, которая будет возвращать сколько дней
прошло или осталось до заданной даты в году, в зависимости от
того, была уже эта дата или нет.

```python
import datetime

def how_much_days(date_str):
    target_date = datetime.datetime.strptime(date_str, '%d-%m-%Y')
    today = datetime.datetime.now()

    if target_date.year != today.year:
        raise ValueError("Дата должна быть в текущем году")
    diff = (target_date - today).days
    if diff > 0:
        return f"До {target_date.date()} осталось {diff} дней"
    elif diff < 0:
        return f"{abs(diff)} дней(дня) прошло с {target_date.date()}"
    else:
        return "Сегодня указанная дата"

date_input = '31-08-2024'
result = how_much_days(date_input)
print(result)
```

7. Сделайте функцию, которая будет возвращать сколько дней
осталось до конца текущего месяца.

```python
import datetime

def days_until():
    today = datetime.date.today()
    next_month = today.replace(day=28) + datetime.timedelta(days=4)  # так осуществляем переход к следующему месяцу
    last_day = next_month - datetime.timedelta(days=next_month.day)
    return (last_day - today).days

print(days_until())
```

9. Напишите код, которой переделает структуру данных

```python
events = [
    {'date': '2019-12', 'event': 'name1'},
    {'date': '2019-12', 'event': 'name2'},
    {'date': '2019-11', 'event': 'name3'},
    {'date': '2019-11', 'event': 'name4'},
    {'date': '2020-10', 'event': 'name5'},
    {'date': '2020-10', 'event': 'name6'},
    {'date': '2020-11', 'event': 'name5'},
    {'date': '2020-11', 'event': 'name6'},
    {'date': '2020-12', 'event': 'name7'},
    {'date': '2020-12', 'event': 'name8'},
    {'date': '2020-12', 'event': 'name9'},
]

def remake(events):
    result = {}
    for event in events:
        year, month = map(int, event['date'].split('-')) # map в данном случае преобразовывает строки в целые числа
        if year not in result:
            result[year] = {} # если текущий год не существует в словаре result, создаем для него новый пустой словарь.
        if month not in result[year]:
            result[year][month] = [] # если месяц не существует в подсловаре для текущего года, создаем для него новый пустой список.
        result[year][month].append(event['event'])
    return result

structured_events = remake(events)
print(structured_events)
```

12. Дан некоторый URL:
$url = 'http://test.com/dir1/dir2/dir3/page.html';
Получите из него последовательный набор папок:
[
'/dir1/dir2/dir3/',
'/dir2/dir3/',
'/dir3/'
]

```python
from urllib.parse import urlparse

url = 'http://test.com/dir1/dir2/dir3/page.html'
path = urlparse(url).path # извлекаем путь из URL
directories = path.strip('/').split('/') # разбиваем путь на массив папок

folders = []
current_path = ''

for i in range(len(directories) - 1):
    current_path += '/' + directories[i]
    folders.append(current_path + '/') # формируем последовательный набор папок

print(folders)
```

13. Сделайте функцию, которая будет устанавливать правильную
форму существительного после числа.

```python
def foo(number, singular, plural, plural_many):
    last_digit = number % 10
    last_two_digits = number % 100
    if last_two_digits in (11, 12, 13, 14):  # исключения для 11-14
        form = plural_many
    elif last_digit == 1:
        form = singular
    elif last_digit in (2, 3, 4):
        form = plural
    else:
        form = plural_many

    return f"{number} {form}"

print(foo(1, 'яблоко', 'яблока', 'яблок'))
print(foo(2, 'яблоко', 'яблока', 'яблок'))
print(foo(3, 'яблоко', 'яблока', 'яблок'))
print(foo(4, 'яблоко', 'яблока', 'яблок'))
print(foo(5, 'яблоко', 'яблока', 'яблок'))
print(foo(11, 'яблоко', 'яблока', 'яблок'))
print(foo(12, 'яблоко', 'яблока', 'яблок'))
print(foo(21, 'яблоко', 'яблока', 'яблок'))
print(foo(23, 'яблоко', 'яблока', 'яблок'))
print(foo(1223421, 'яблоко', 'яблока', 'яблок'))
```

14. Даны два числа. Выведите в консоль процесс умножения этих
чисел в столбик, как в школе.

```python
def multiplication(num1, num2):
    str_num1 = str(num1)
    str_num2 = str(num2)
    length_num2 = len(str_num2)

    print(f"  {str_num1}") # тут печатаем верхнее число
    print(f"x {str_num2}") # тут печатаем верхнее число
    print('-' * (length_num2 + 2))  # отрисовываем линию для разделения
    results = [] # в этой переменной хранится промежуточный результат

    # само умножение
    for i in range(length_num2):
        digit2 = int(str_num2[length_num2 - 1 - i])  # текущая цифра второго числа
        product = num1 * digit2
        print(f"{' ' * (i + 1)}{product}")  # печатаем промежуточный результат
        results.append(product * (10 ** i))  # сохраняем промежуточные результаты в ранее подготовленный список(изначально пустой)
    print('-' * (length_num2 + 2))  # отрисовываем линию для разделения

    # получаем и выводим итоговое значение
    total = sum(results)
    print(f"  {total}")

multiplication(33, 99)
```

15. Даны два числа. Выведите на экран процесс деления этих чисел
в столбик, как в школе.

```python
def division(dividend, divisor):
    str_dividend = str(dividend)
    str_divisor = str(divisor)
    length_dividend = len(str_dividend)
    print(f" {str_dividend}") # печатаем верхнее число
    print(f"÷ {str_divisor}") # печатаем верхнее число
    print('-' * (length_dividend + 2))  # отрисовываем линию разделения
    quotient = ""
    remainder = 0

    # деление
    for i in range(length_dividend):
        # берем следующую цифру из делимого
        current = remainder * 10 + int(str_dividend[i])
        if current < divisor:
            quotient += "0"  # добавляем 0 в частное
            if i == 0:
                print(" " * (i + 2) + "0")  # если первая цифра меньше делителя, выводим 0
            continue

        current_quotient = current // divisor
        remainder = current % divisor # находим частное
        quotient += str(current_quotient) # находим остаток
        print(" " * (i + 1) + str(current))  # печатаем текущее делимое
        print(" " * (i + 1) + str(current_quotient * divisor))  # печатаем вычитаемое
        print(" " * (i + 1) + "-" * len(str(current_quotient * divisor)))  # отрисовка линии разделения
        print(" " * (i + 1) + str(remainder))  # печатаем остаток

    print(f"Частное: {quotient} с остатком {remainder}")

division(555, 4)
```

16. Дано число. Выведите на экран процесс разложения числа на
простые множители, как в школе.

```python
def decompose(n):
    factors = []
    divisor = 2  # начинаем деление с 2

    while n > 1:
        while n % divisor == 0:  # если n делится на divisor, то делим и выводим процесс
            factors.append(divisor)
            n //= divisor
            print(f"{n * divisor} ÷ {divisor} = {n} (множитель: {divisor})")  #вывод процесса
        divisor += 1  # переходим к следующему числу

    return factors

number = 80
print(f"Разложение числа {number} на простые множители:")
factors = decompose(number)
print("Простые множители:", factors)
```

17. Напишите программу, которая будет находить корни квадратного
уравнения. Для этого спросите у пользователя три числа, которые
будут коэффициентами уравнения.

```python
import math
def solution(a, b, c):
    D = b**2 - 4*a*c
    if D > 0: # два различных корня
        root1 = (-b + math.sqrt(D)) / (2 * a)
        root2 = (-b - math.sqrt(D)) / (2 * a)
        return (root1, root2)
    elif D == 0: # один корень
        root = -b / (2 * a)
        return (root,)
    else: # корней нет
        return None

try:
    a = float(input("Введите коэффициент a (не должен быть равен 0): "))
    if a == 0:
        raise ValueError("Коэффициент a не может быть равен 0.")
    b = float(input("Введите коэффициент b: "))
    c = float(input("Введите коэффициент c: "))
    roots = solution(a, b, c) # находим корни

    if roots is None:
        print("Корней нет (дискриминант меньше 0).")
    elif len(roots) == 1:
        print(f"Уравнение имеет один корень: {roots[0]}")
    else:
        print(f"Уравнение имеет два корня: {roots[0]} и {roots[1]}")
except ValueError:
    print(f"Ошибка: {ValueError}")
```

18. Напишите код, который будет генерировать пароль заданного
размера. Пароль обязательно должен будет содержать маленькую
и большую буквы, цифру и какой-то специальный символ.

```python
import random
import string

def generate_password(length):
    if length < 4:
        raise ValueError("Длина пароля должна быть не менее 4 символов.") # исходя из условия задачи

    lowercase = string.ascii_lowercase # наборы символов получаем, используя библиотеку string
    uppercase = string.ascii_uppercase
    digits = string.digits
    special_characters = string.punctuation
    password = [ # выполняем условие задачи (что в пароле есть цифра, символ, большая и маленькая буквы)
        random.choice(lowercase),
        random.choice(uppercase),
        random.choice(digits),
        random.choice(special_characters)
    ]
    all_characters = lowercase + uppercase + digits + special_characters
    password += random.choices(all_characters, k=length - 4) # заполняем оставшуюся длину пароля случайными символами из всех наборов
    random.shuffle(password) # на всякий его можно еще и перемешать
    return ''.join(password)

try:
    length = int(input("Введите длину пароля (минимум 4): "))
    password = generate_password(length)
    print(f"Сгенерированный пароль: {password}")
except ValueError:
    print(f"Ошибка: {ValueError}")
```

19. Напишите код, который будет проверять сложность пароля.
Критерии сложности придумайте самостоятельно.

```python
import string

def check_password(password):
    min_length = 6 # укажем мин. длину
    has_upper = any(c.isupper() for c in password) # пусть будет минимум одна большая буква
    has_lower = any(c.islower() for c in password) # и маленькая
    has_digit = any(c.isdigit() for c in password) # и циферка пусть будет
    has_special = any(c in string.punctuation for c in password) # и спец. символ
    is_long_enough = len(password) >= min_length

    if all([has_upper, has_lower, has_digit, has_special, is_long_enough]): # проверяем условия
        return "Пароль шикарный."
    else:
        return "Так себе пароль."

password = input("Введите пароль для проверки сложности: ")
result = check_password(password)
print(result)
```

20. Реализуйте алгоритм Решето Эратосфена для поиска простых
чисел в заданном промежутке.

```python
def eratosfen(limit):

    is_ez = [True] * (limit + 1) # создаем список, где индекс будет представлять число, а значение - True/False
    is_ez[0], is_ez[1] = False, False  # 0 и 1 не являются простыми числами
    for number in range(2, int(limit**0.5) + 1):
        if is_ez[number]:  # если число простое, удаляем все кратные числа
            for multiple in range(number * number, limit + 1, number):
                is_ez[multiple] = False
    return [num for num, ez in enumerate(is_ez) if ez] # возвращаем список простых чисел

try:
    lower_limit = int(input("Введите нижнюю границу диапазона: "))
    upper_limit = int(input("Введите верхнюю границу диапазона: "))

    if lower_limit < 0 or upper_limit < 0 or lower_limit > upper_limit:
        raise ValueError("Границы должны быть неотрицательными и нижняя граница не должна превышать верхнюю.")

    ezes = eratosfen(upper_limit)
    ezes_in_range = [p for p in ezes if p >= lower_limit]

    print(f"Простые числа в диапазоне от {lower_limit} до {upper_limit}: {ezes_in_range}")

except ValueError:
    print(f"Ошибка: {ValueError}")
```
