
    Сортировка 1

    Сгенерировать два списка x и y
    Записать в json файл данные из списков x и y
    Отсортировать данные из json файла методом пузырьком (по возрастанию)
    Построить график по итоговым данным

    Сортировка 2

    Сгенерировать два списка x и y
    Записать в json файл данные из списков x и y
    Отсортировать данные из json файла методом пузырьком (по убыванию)
    Построить график по итоговым данным

    Сортировка 3

    Сгенерировать два списка x и y
    Записать в json файл данные из списков x и y
    Отсортировать данные из json файла методом выбора (по возрастанию)
    Построить график по итоговым данным


```python
import json
import random
import matplotlib.pyplot as plt

x = [random.randint(1, 100) for i in range(10)] # генерю списки
y = [random.randint(1, 100) for i in range(10)]

data = x + y

with open('data.json', 'w') as f:
    json.dump(data, f)

# пузырьковая сортировка по возрастанию
def bubble_sort_asc(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n-i-1):
            if data[j] > data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]
    return data

# пузырьковая сортировка по убыванию
def bubble_sort_desc(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n-i-1):
            if data[j] < data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]
    return data

# сортировка методом выбора по возрастанию
def selection_sort_asc(data):
    n = len(data)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if data[j] < data[min_idx]:
                min_idx = j
        data[i], data[min_idx] = data[min_idx], data[i]
    return data

data_sorted = bubble_sort_asc(data.copy()) # пузырьковая сортировка по возрастанию

plt.figure(figsize=(8, 6))
plt.bar(range(len(data_sorted)), data_sorted, color='b')
plt.title("Сортировка 1: Пузырьковая сортировка по возрастанию")
plt.xlabel('Индекс')
plt.ylabel('Значение')

# устанавливаю границы оси X от 0 до 20 с шагом 1
plt.xlim(0, 20)
plt.xticks(range(21))

plt.grid(True)
plt.show()

data_sorted = bubble_sort_desc(data.copy()) #пузырьковая сортировка по убыванию

plt.figure(figsize=(8, 6))
plt.bar(range(len(data_sorted)), data_sorted, color='g')
plt.title("Сортировка 2: Пузырьковая сортировка по убыванию")
plt.xlabel('Индекс')
plt.ylabel('Значение')

# устанавливаю границы оси X от 0 до 20 с шагом 1
plt.xlim(0, 20)
plt.xticks(range(21))

plt.grid(True)
plt.show()

# сортировка методом выбора по возрастанию
data_sorted = selection_sort_asc(data.copy())

plt.figure(figsize=(8, 6))
plt.bar(range(len(data_sorted)), data_sorted, color='r')
plt.title("Сортировка 3: Сортировка методом выбора по возрастанию")
plt.xlabel('Индекс')
plt.ylabel('Значение')

# устанавливаю границы оси X от 0 до 20 с шагом 1
plt.xlim(0, 20)
plt.xticks(range(21))

plt.grid(True)
plt.show()
```
