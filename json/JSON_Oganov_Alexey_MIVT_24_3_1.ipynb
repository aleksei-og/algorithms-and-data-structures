**1.** Программа получает навход описание одного объекта в формате JSON.
 Программа должнавывести все парыключ–значение этого объекта,
 разделяя их двоеточием. Используйте для этого функцию json.loads.
 Выведите пары ключ–значение, разделяя их двоеточием и пробелом,
 упорядочив по ключу. Кавычки вокруг строк выводить не надо, см. пример.

```python
import json

string = '{"street": "Большая Дмитровка","city": "Москва","country": "Россия"}'
data = json.loads(string)

sorted_data = dict(sorted(data.items())) # сортировка

result = '\n'.join([f'{key}: {value}' for key, value in sorted_data.items()]) # преобразование в строку заданного формата
print(result)
```

**3.** Объекты в JSON могут быть следующих типов
 Строки (рекомендуется представление в кодировке UTF-8). Например,
 "Hello".
 Числа (целые иливещественные). Например, 179 или 1.79.
 Логические (true или false).
 null (нулевое значение).
 Массивы (списки), задаются в квадратных скобках. Например, [1, 2, 3].
 Объекты JSON(словари), задаются в фигурных скобках. Например, {"city":
 "Moscow"}.
 ВамданJSON-объект, являющийся списком. Пройдитесь по его элементам,
 исовсемиэлементами сделайте следующее:
 Если это строка, добавьте к конец строки восклицательный знак.
 Если это число, то увеличьте его на 1.
 Если это логическое значение, то инвертируйте его.
 Если это список, то удвойте его.
 Если это объект, то добавьте к нему новый атрибут с ключом "newkey" и
 пустым значением.
 Если это пустое значение, то удалите его из списка.
 Выведите полученный JSON

```python
import json

json_list = ["Hello", 179, 0.5, True, [1,2,3], {"key":"value"}, None]

for i in range(len(json_list)):
  if isinstance(json_list[i], bool):
    json_list[i] = not json_list[i]
  elif isinstance(json_list[i], str):
    json_list[i] += '!'
  elif isinstance(json_list[i], (int, float)):
    json_list[i] += 1
  elif isinstance(json_list[i], list):
    json_list[i] *= 2
  elif isinstance(json_list[i], dict):
    key = 'newkey'
    json_list[i][key] = None
  elif json_list[i] == None:
    del json_list[i]

print(json.dumps(json_list, ensure_ascii=False))
```

**4.** Даны два объекта JSON.Необходимообъединитьихвместепоследующему
 правилу:результатберётвсеатрибутыизпервогообъекта,аесливовтором
 объекте содержатся атрибуты с такимиже ключами, как и у первого
 объекта,илиотсутствующиевпервомобъекте,тоонитакжедобавляютсяв
 результат.
 ПрограммаполучаетнавходJSON-списокиздвухJSON-объектовидолжна
 вывестиодинJSON-объект

Первый вариант решения, так сказать "в лоб"

```python
import json

json_string = '[{"firstName":"Иван","middleName":"Петров","birthDate":"2001/02/29"}, {"city":"Moscow","birthDate":"2001/02/28"}]'
json_object = json.loads(json_string)

first_dict = json_object[0]
second_dict = json_object[1]

result_list = []

for i in first_dict.items():
  for j in second_dict.items():
    if i[0] in second_dict.keys():
      result_list.append(j)
    else:
      result_list.append(i)

result_list = list(dict.fromkeys(result_list)) # удаляем дубликаты из списка
result_json = dict(result_list)
output_json = json.dumps(result_json, ensure_ascii=False)
print(output_json)
```

Второй вариант решения, используя существующие методы

```python
import json

data = [
    {"firstName": "Иван", "middleName": "Петров", "birthDate": "2001/02/29"},
    {"city": "Moscow", "birthDate": "2001/02/28"}
]

first_object = data[0]
second_object = data[1]

result = first_object.copy()  # копирую первый словарь
result.update(second_object)  # обновляю результат значениями из второго словаря

output_json = json.dumps(result, ensure_ascii=False)
print(output_json)
```

```python
import json

objects = [{"firstName":"Иван","middleName":
 "Петров","birthDate":"2001/02/29"},
 {"firstName":"Пётр","middleName":
 "Иванов","city":"Moscow"}]

all_keys = set()
for obj in objects: # собираю уникальные ключи
    all_keys.update(obj.keys())

for obj in objects: # добавляю недостающие атрибуты
    for key in all_keys:
        if key not in obj:
            obj[key] = None

output_json = json.dumps(objects, ensure_ascii=False)
print(output_json)
```

**1.** Получите файл у преподавателя и обработайте его.
 Посчитайте для каждого района Москвы количество точек доступа в
 этом районе. Упорядочить список по убыванию количества точек
 доступа, при равном значении — в лексикографическом порядке
 названия района. Выведите полученный список, отделяя название
 района от числа точек доступа двоеточием и пробелом

```python
import json
from collections import defaultdict

def count_access_points(json_file):
    with open(json_file, 'r', encoding='Windows-1251') as file:
        data = json.load(file)

    district_counts = defaultdict(int) # это словарь для подсчета точек доступа по районам

    for entry in data: # пробегаюсь по данным и суммирую кол-во точек доступа
        district = entry['District']
        num_access_points = entry['NumberOfAccessPoints']
        district_counts[district] += num_access_points

    sorted_districts = sorted(district_counts.items(), key=lambda x: (-x[1], x[0])) # преобразую словарь в список кортежей (район, количество)

    for district, count in sorted_districts: # преобразую районы, если нужно, например, убрать или добавить "район" в название
        formatted_district = district
        if district.startswith('район '):
            formatted_district = district[6:]  # убираю "район ", используя срез

        print(f"{formatted_district}: {count}")

count_access_points('wifi.json')
```

**2.** Получите файл у преподавателя и обработайте его.
 Вамнеобходимо посчитать для каждой московской школы
 количество победителей и призёров регионального этапа
 всероссийской олимпиады школьников (региональный этап — это
 третий этап всероссийской олимпиады).
 Необходимо вывести список всех школы Москвывформате:краткое
 название школы, затем через двоеточие и пробел количество
 победителей регионального этапа из этой школы, количество
 призёров регионального этапа, через пробел.
 Список упорядочен по убыванию суммарного числа победителей и
 призёров, затем по числу победителей, затем по названию школы

**Примечание**
В файле olympics.json сделал исправление в столбце 32506035 - убрал лишнюю кавычку, иначе выдает ошибку.

```python
import json
from collections import defaultdict

def count_olympiad_winners_and_prizewinners(json_file):

    with open(json_file, 'r', encoding='Windows-1251') as file:
        data = json.load(file)

    school_results = defaultdict(lambda: {'winners': 0, 'prizewinners': 0}) # словарь для подсчета победителей и призеров по школам

    for entry in data: # прохожу по данным и подсчитываю победителей и призёров для каждой школы
        if entry['Stage'] == '3':  # рассматриваю только региональный этап
            school = entry['ShortName']
            status = entry['Status']

            if status == 'победитель':
                school_results[school]['winners'] += 1
            elif status == 'призёр':
                school_results[school]['prizewinners'] += 1

    # сортируем результаты:
    # 1. по убыванию общего числа (winners + prizewinners)
    # 2. по количеству победителей
    # 3. лексикографически по названию школы
    sorted_results = sorted(school_results.items(),
                            key=lambda item: (-item[1]['winners'] - item[1]['prizewinners'],
                                              -item[1]['winners'],
                                              item[0]))

    for school, results in sorted_results: # вывод в заданном формате
        total = results['winners'] + results['prizewinners']
        print(f"{school}: {results['winners']} {results['prizewinners']}")

count_olympiad_winners_and_prizewinners('olympics.json')
```

**3.** Определите, по каким предметам проводится московская олимпиада.
 Длякаждого предмета введите его название, и сколько лет
 проводится олимпиада по этому предмету.
Выведите список предметов в алфавитном порядке, затем через
 двоеточие и пробел —количество лет проведения.
 Начальная часть ответа выглядит так:
 Биология: 7
 Бюджетная грамотность: 1
 География: 7
 Изобразительное искусство: 7
 Иностранный язык(испанский язык):

```python
import json
from collections import defaultdict

def count_olympiad_years(json_file):

    with open(json_file, 'r', encoding='Windows-1251') as file:
        data = json.load(file)

    subject_years = defaultdict(set) # словарь для подсчета лет проведения олимпиады по предметам

    for entry in data: # прохожу по данным и собираю года проведения олимпиады по каждому предмету
        olympiad_name = entry['OlympiadName']
        year = entry['Year']

        subject_years[olympiad_name].add(year) # добавляю год в множество для каждого предмета


    sorted_subjects = sorted(subject_years.items(), key=lambda x: x[0])  # сортирую по названию предмета

    for subject, years in sorted_subjects: # вывод в нужном формате
        print(f"{subject}: {len(years)}")

count_olympiad_years('olympics.json')
```
